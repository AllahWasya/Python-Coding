```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filename="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
df=pd.read_csv(filename)

```

```python
headers= ["symboling","normalized-losses","make","fuel-type", "aspiration","num of doors","body-style"
        ,"drive-wheels","engin-location","Wheel-base", "length", "width", "height","cur-weight","engine-type",
        "nub of cylinder","engine size", "fuel-system", "bore","stroke", "compression ratio", "horse power"
        ,"peak-rpm","city-mpg","high-way(mpg)","price" ]
```

```python
df=pd.read_csv(filename, names= headers)
df
```

> ## Replacing ? with NaN


```python
df.replace('?', np.nan, inplace=True)
df
```

> ## Missing Data 

```python
Missing_data=df.isnull()
Missing_data
```

```python
check_missing=df.isnull()
for columns in check_missing.columns.values.tolist():
    print(columns)
    print(check_missing[columns].value_counts())
```

```python
avg_nor_loss=df["normalized-losses"].astype("float").mean(axis=0)
print("avg_normalized Loss: ", avg_nor_loss)
df["normalized-losses"].replace(np.nan,avg_nor_loss,inplace=True)
```

```python
avg_nor_loss=df["bore"].astype("float").mean(axis=0)
print("Bore avg value: ", avg_nor_loss)
df["bore"].replace(np.nan,avg_nor_loss,inplace=True)
```

```python
avg_nor_loss=df["horse power"].astype("float").mean(axis=0)
print("avg_normalized Loss: ", avg_nor_loss)
df["horse power"].replace(np.nan,avg_nor_loss,inplace=True)
```

> ## Now insted of mean 
> ## Replacing with high frequency value

```python
x=df["num of doors"].value_counts().idxmax()
df["num of doors"].replace(np.nan, x, inplace=True)
```

> ## Drop Whole row where price is NaN


```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.head()
```

```python
# Droping three rows 204x26 to 201x26
df.shape
```

```python
df.dtypes
```

```python
df[["bore","stroke"]]=df[["bore","stroke"]].astype("float")
df[["normalized-losses"]]=df[["normalized-losses"]].astype("int")
```

```python
df[["price"]]=df[["price"]].astype("float")
df[["peak-rpm"]]=df[["peak-rpm"]].astype("float")
```

> ## Convert Miles per hour to Litre/km
> ## Add a cloumns of it


```python
df["city-L/km"]=245/df["city-mpg"]
```

## Normalizing Length and Width



```python
df["length"]=df["length"]/df["length"].max()
df["width"]=df["width"]/df["width"].max()
```

## Binning

```python
df["horse power"]=df["horse power"].astype("int",copy=True)
bins=np.linspace(min(df["horse power"]),max(df["horse power"]),4)
goup_names=["low","medium","high"]
df["horse_power_binned"]=pd.cut(df["horse power"],bins,labels=goup_names,include_lowest=True)
df["horse_power_binned"]
df[["horse power","horse_power_binned"]].head(20)
```

```python
df["horse_power_binned"].value_counts()
```

```python
import seaborn as sns
sns.barplot(x=goup_names,y=df["horse_power_binned"].value_counts())
plt.xlabel("Horsepowern (HP)")
plt.ylabel("count")
plt.show()
```
